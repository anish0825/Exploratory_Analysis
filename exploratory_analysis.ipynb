```python
def by_month(text):
    if text[5:7]=='01':
        return 'January'
    elif text[5:7]=='02':
        return 'February'
    elif text[5:7]=='03':
        return 'March'
    elif text[5:7]=='04':
        return 'April'
    elif text[5:7]=='05':
        return 'May'
    elif text[5:7]=='06':
        return 'June'
    elif text[5:7]=='07':
        return 'July'
    elif text[5:7]=='08':
        return 'August'
    elif text[5:7]=='09':
        return 'September'
    elif text[5:7]=='10':
        return 'October'
    elif text[5:7]=='11':
        return 'November'
    else:
        return 'December'
```
